```python
from sklearn.model_selection import train_test_split
```

```python
df_train,df_test=train_test_split(df_imp, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df_imp
```

```python
num_vars = ['wheelbase', 'curbweight', 'enginesize', 'boreratio', 'horsepower','fuelEconomy','carlength','carwidth','price']
```

```python
df_train.loc[:,num_vars]=scaler.fit_transform(df_train[num_vars])
```

```python
df_train
```

```python
y_train=df_train.pop("price")
X_train=df_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE 
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
rfe=RFE(lm,10)
rfe=rfe.fit(X_train,y_train)
```

```python
df_train.info()
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
X_train.columns[rfe.support_]
```

```python
X_train_rfe=X_train[X_train.columns[rfe.support_]]
X_train_rfe
```

```python
def build_model(X,y):
    X=sm.add_constant(X)
    lm=sm.OLS(y,X).fit()
    print(lm.summary())
    return X
```

```python
def checkVIF(X):
    vif=pd.DataFrame()
    vif["Features"]=X.columns
    vif["VIF"]=(variance_inflation_factor(X.values,i) for i in range(X.shape[1]))
    vif["VIF"]=round(vif["VIF"],2)
    vif=vif.sort_values(by="VIF",ascending=False)
    return(vif)
```

```python
X_train_new=build_model(X_train_rfe,y_train)
```

```python
X_train_new= X_train_rfe.drop(["carwidth"],axis=1)
```

```python
X_train_new=build_model(X_train_new,y_train)
```

```python
checkVIF(X_train_new)
```

```python
def checkVIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

```python
X_train_new = X_train_new.drop(["enginesize"], axis = 1)
```

```python
X_train_new = build_model(X_train_new,y_train)
```

```python
X_train_new=X_train_new.drop(["cylindernumber_twelve"],axis=1)
```

```python
checkVIF(X_train_new)
```

```python
X_train_new = build_model(X_train_new,y_train)
```

```python
lm = sm.OLS(y_train,X_train_new).fit()
y_train_price = lm.predict(X_train_new)
```

```python
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)
```

```python
num_vars = ['wheelbase', 'curbweight', 'enginesize', 'boreratio', 'horsepower','fuelEconomy','carlength','carwidth','price']
df_test[num_vars] = scaler.fit_transform(df_test[num_vars])
```

```python
y_test = df_test.pop('price')
X_test = df_test
```

```python
# Now let's use our model to make predictions.
X_train_new = X_train_new.drop('const',axis=1)
# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_new.columns]

# Adding a constant variable 
X_test_new = sm.add_constant(X_test_new)
```
